```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2  # Use OpenCV for image loading

def load_images(directory):
    return [cv2.imread(os.path.join(directory, img)) for img in os.listdir(directory)]

def display_images(images, labels, title):
    fig, axes = plt.subplots(1, len(images), figsize=(20, 8))
    for i in range(len(images)):
        axes[i].imshow(images[i])
        axes[i].set_title(labels[i])
    plt.suptitle(title)
    plt.show()

def plot_distribution(data, labels, title):
    plt.bar(labels, data, color='skyblue')
    plt.title(title)
    plt.show()

# Load and display sample images
sample_images = []
sample_labels = []
for emotion in target_var:
    emotion_dir = os.path.join(train_dir, emotion)
    sample_image_path = os.path.join(emotion_dir, os.listdir(emotion_dir)[0])
    sample_images.append(cv2.imread(sample_image_path))
    sample_labels.append(emotion)

display_images(sample_images, sample_labels, "Sample Images")


```

```python
# Plot distribution of images
train_counts = np.array([len(os.listdir(os.path.join(train_dir, i))) for i in target_var])
test_counts = np.array([len(os.listdir(os.path.join(test_dir, i))) for i in target_var])

plot_distribution(train_counts, target_var, "Training Set Distribution")
plot_distribution(test_counts, target_var, "Testing Set Distribution")

# Print dataset statistics
for emotion in target_var:
    print(f'Emotion: {emotion}')
    print(f'\tTraining: {len(os.listdir(os.path.join(train_dir, emotion)))}')
    print(f'\tTesting: {len(os.listdir(os.path.join(test_dir, emotion)))}')
```

```python
import random

def display_random_images(images, labels, title, num_samples=5):
    random_indices = random.sample(range(len(images)), num_samples)
    selected_images = [images[i] for i in random_indices]
    selected_labels = [labels[i] for i in random_indices]

    display_images(selected_images, selected_labels, title)

# Display random images
display_random_images(sample_images, sample_labels, "Random Image Samples")
```

```python

def plot_image_dimensions(images, labels, title):
    dimensions = [(img.shape[0], img.shape[1]) for img in images]
    widths, heights = zip(*dimensions)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].hist(widths, bins=20, color='skyblue', alpha=0.7)
    axes[0].set_title('Width Distribution')
    axes[1].hist(heights, bins=20, color='salmon', alpha=0.7)
    axes[1].set_title('Height Distribution')
    plt.suptitle(title)
    plt.show()

# Display image dimension distributions
plot_image_dimensions(sample_images, sample_labels, "Image Dimensions Distribution")

```

```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Function to load and preprocess images
def load_and_preprocess_data(directory):
    images = []
    labels = []
    emotion_labels = os.listdir(directory)

    for label, emotion in enumerate(emotion_labels):
        emotion_dir = os.path.join(directory, emotion)
        for img_name in os.listdir(emotion_dir):
            img_path = os.path.join(emotion_dir, img_name)
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
            img = cv2.resize(img, (48, 48))  # Resize images to a common size
            images.append(img)
            labels.append(label)

    return np.array(images), np.array(labels)

# Load and preprocess the data
root_dir = 'C:\\Users\\chand\\Desktop\\FinalYearProject\\arch'
train_dir = os.path.join(root_dir, 'train')
test_dir = os.path.join(root_dir, 'test')

X_train, y_train = load_and_preprocess_data(train_dir)
X_test, y_test = load_and_preprocess_data(test_dir)

# Normalize pixel values to be between 0 and 1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode the labels
num_classes = len(os.listdir(train_dir))
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

# Split the training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train.reshape(-1, 48, 48, 1), y_train, epochs=30, batch_size=32, validation_data=(X_val.reshape(-1, 48, 48, 1), y_val))

# Evaluate the modQel on the test set
test_loss, test_accuracy = model.evaluate(X_test.reshape(-1, 48, 48, 1), y_test)
print(f'Test Accuracy: {test_accuracy}')

# Visualize training history
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
